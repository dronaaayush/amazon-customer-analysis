```python
#Gender Distribution 
import matplotlib.pyplot as plt
gender_distribution = df['Gender'].value_counts()
gender_distribution.plot(kind='pie', figsize=(8,9), autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Age Distribution
import seaborn as sns
import matplotlib.pyplot as plt

age_distribution = df['age'].describe()

ax = sns.histplot(df['age'], kde=True, bins=20)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                fontsize=10, color='blue', 
                xytext=(0, 5), textcoords='offset points' )

plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#What is the correlation between "Purchase_Frequency" and "Purchase_Categories"?
correlation_data = pd.crosstab(df['Purchase_Frequency'], ['Purchase_categories'])

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_data, annot=True, fmt='d', cmap='Blues', cbar=False)

plt.title('Heatmap of Purchase Frequency vs Purchase Categories')
plt.xlabel('Purchase Categories')
plt.ylabel('Purchase Frequency')
plt.show()
```

```python
#Do certain product categories have a higher likelihood of abandonment in the shopping cart?
df.groupby(['Purchase_Categories', 'Cart_Abandonment_Factors']).size().unstack().plot(kind='bar', figsize=(19, 9))
plt.show()
```

```python
#How does "Browsing_Frequency" impact "Add_to_Cart_Browsing" behavior?
df.groupby(['Browsing_Frequency', 'Add_to_Cart_Browsing']).size().unstack().plot(kind='bar', figsize=(10,12))
plt.show()
```

```python
#Is there a relationship between the "Product_Search_Method" and "Search_Result_Exploration" behavior?
relationship_corr = df.groupby('Product_Search_Method')['Search_Result_Exploration'].size()
relationship_corr.plot(figsize=(12,10), kind='bar')
plt.show()

```

```python
#How do customer reviews ("Review_Helpfulness", "Review_Reliability") influence "Shopping_Satisfaction"
#and "Service_Appreciation"?

import seaborn as sns
import matplotlib.pyplot as plt

aggregated_data = df.groupby('Review_Helpfulness')['Shopping_Satisfaction'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='Review_Helpfulness', y='Shopping_Satisfaction', data=aggregated_data)
plt.title('Shopping Satisfaction by Review Helpfulness')
plt.xlabel('Review Helpfulness')
plt.ylabel('Average Shopping Satisfaction')
plt.show()

```

```python
#How effective are personalized recommendations based on the "Personalized_Recommendation_Frequency" in increasing
#purchase frequency or satisfaction?
recom_freq = df.groupby('Personalized_Recommendation_Frequency')['Purchase_Frequency'].count()
recom_freq.plot(kind='bar', figsize=(9,9))
plt.title('Personal Recommendation Frequency in Increasing Purchase Frequency')
plt.ylabel('Purchase Frequency')
plt.xlabel('Personal Recommendation')
plt.show()

```

```python
#"Personalized_Recommendation_Frequency" in increasing satisfaction?
satis_freq = df.groupby('Personalized_Recommendation_Frequency')['Shopping_Satisfaction'].mean()
satis_freq.plot(kind='bar', figsize=(9,9))
plt.title('Personalized Recommendation Frequency in increasing satisfaction')
plt.xlabel('Personalized Recommendation')
plt.ylabel('Shopping Satisfaction')
plt.show()
```

```python
#What are the main factors contributing to cart abandonment ("Cart_Abandonment_Factors")?
cart_abandonment_factors = df['Cart_Abandonment_Factors'].value_counts()
cart_abandonment_factors.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))
plt.title('Factors Contributing to Cart Abandonment')
plt.ylabel('')
plt.show()
```

```python
#cart_abandonment_by_category 

cart_abandonment_by_category = df.groupby('Purchase_Categories')['Cart_Abandonment_Factors'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=cart_abandonment_factors.index, y=cart_abandonment_factors.values, palette='viridis')
plt.title('Frequency of Cart Abandonment Factors')
plt.xlabel('Cart Abandonment Factors')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
cart_abandonment_by_category

```

```python
#How do different "Cart_Abandonment_Factors" correlate with shopping satisfaction and the likelihood of leaving
#a review ("Review_Left")?

# Convert 'Review_Left' to binary (e.g., Yes=1, No=0)
df['Review_Left_Binary'] = df['Review_Left'].apply(lambda x: 1 if x == 'Yes' else 0)

review_by_abandonment = df.groupby('Cart_Abandonment_Factors')['Review_Left_Binary'].mean()

plt.figure(figsize=(12, 6))
sns.barplot(x=review_by_abandonment.index, y=review_by_abandonment.values, palette='magma')
plt.title('Proportion of Reviews Left by Cart Abandonment Factors')
plt.xlabel('Cart Abandonment Factors')
plt.ylabel('Proportion of Reviews Left')
plt.xticks(rotation=45)
plt.show()
```

```python
#How does customer feedback, such as "Improvement_Areas", relate to the overall shopping experience and satisfaction?
satisfaction_by_improvement = df.groupby('Improvement_Areas')['Shopping_Satisfaction'].mean()
improvement_area_counts = df['Improvement_Areas'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = satisfaction_by_improvement.index
y = satisfaction_by_improvement.values

plt.figure(figsize=(19, 16))

colors = plt.cm.viridis(np.linspace(0, 1, len(x)))  
plt.bar(x, y, color=colors)

plt.title('Shopping Satisfaction by Improvement Areas', fontsize=14, weight='bold')
plt.xlabel('Improvement Areas', fontsize=12)
plt.ylabel('Mean Shopping Satisfaction', fontsize=12)

plt.xticks(rotation=90, ha='right', fontsize=10)

plt.tight_layout()  
plt.show()
```

```python
# Bar plot for frequency of Improvement Areas
plt.figure(figsize=(12, 9))
sns.barplot(x=improvement_area_counts.index, y=improvement_area_counts.values, palette='magma')
plt.title('Frequency of Improvement Areas')
plt.xlabel('Improvement Areas')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()  
plt.show()
```
